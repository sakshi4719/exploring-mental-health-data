```python
test_data = pd.read_csv('/kaggle/input/playground-series-s4e11/test.csv')
train_data = pd.read_csv('/kaggle/input/playground-series-s4e11/train.csv')
```

```python
train_data['Have you ever had suicidal thoughts ?'] = train_data['Have you ever had suicidal thoughts ?'].map({'Yes': 1, 'No': 0})
train_data['Family History of Mental Illness'] = train_data['Family History of Mental Illness'].map({'Yes': 1, 'No': 0})
```

```python
train_data['Academic Pressure'] = train_data['Academic Pressure'].fillna(0)
train_data['Work Pressure'] = train_data['Work Pressure'].fillna(0)

test_data['Academic Pressure'] = test_data['Academic Pressure'].fillna(0)
test_data['Work Pressure'] = test_data['Work Pressure'].fillna(0)

train_data['Pressure'] = train_data['Academic Pressure'] + train_data['Work Pressure']
test_data['Pressure'] = test_data['Academic Pressure'] + test_data['Work Pressure']

train_data['Pressure'] = train_data['Pressure'].replace(0, 3)
test_data['Pressure'] = test_data['Pressure'].replace(0, 3)
```

```python
train_data['Study Satisfaction'] = train_data['Study Satisfaction'].fillna(0)
train_data['Job Satisfaction'] = train_data['Work Pressure'].fillna(0)

test_data['Study Satisfaction'] = test_data['Study Satisfaction'].fillna(0)
test_data['Job Satisfaction'] = test_data['Job Satisfaction'].fillna(0)

train_data['Satisfaction'] = train_data['Study Satisfaction'] + train_data['Job Satisfaction']
test_data['Satisfaction'] = test_data['Study Satisfaction'] + test_data['Job Satisfaction']

train_data['Satisfaction'] = train_data['Satisfaction'].replace(0, 3)
test_data['Satisfaction'] = test_data['Satisfaction'].replace(0, 3)
```

```python
train_data['Financial Stress'] = train_data['Financial Stress'].fillna(2.988983)
```

```python
train_data['Gender'] = train_data['Gender'].map({'Male': 0, 'Female': 1})
test_data['Gender'] = test_data['Gender'].map({'Male': 0, 'Female': 1})
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Define the columns you want to encode
object_cols = ['Dietary Habits', 'Sleep Duration']

# Initialize OrdinalEncoder with handling for unknown categories
ordinal_encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

# Fit the encoder on the training data and transform
train_data[object_cols] = ordinal_encoder.fit_transform(train_data[object_cols])

# Transform the test data based on the fitted encoder
test_data[object_cols] = ordinal_encoder.transform(test_data[object_cols])

# Check the transformed data
print(train_data[object_cols].head())
print(test_data[object_cols].head())
```

```python
train_data['Sleep Duration'] = train_data['Sleep Duration'].fillna(3)
test_data['Sleep Duration'] = test_data['Sleep Duration'].fillna(3)
```

```python
train_data['Dietary Habits'] = train_data['Dietary Habits'].fillna(2)
test_data['Dietary Habits'] = test_data['Dietary Habits'].fillna(2)
```

```python
train_data['Work-Life Balance'] = train_data['Work/Study Hours'] / train_data['Sleep Duration'].replace(0, 1)
test_data['Work-Life Balance'] = test_data['Work/Study Hours'] / test_data['Sleep Duration'].replace(0, 1)
```

```python
features = ['Gender', 'Age', 'Pressure', 'Satisfaction', 'Work/Study Hours', 'Financial Stress', 'Sleep Duration', 'Dietary Habits', 
        'Work-Life Balance']
```

```python
X_train = train_data[features]
X_test = test_data[features]

y = train_data.Depression
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.callbacks import ReduceLROnPlateau

model = keras.Sequential([
    # First hidden layer
    layers.Dense(units=512, activation='relu', input_shape=[X_train.shape[1]]),
    layers.Dropout(0.5),  # Optional: Add dropout to reduce overfitting
    layers.BatchNormalization(),  # Optional: Add batch normalization
    
    # Second hidden layer
    layers.Dense(units=512, activation='relu'),
    layers.Dropout(0.5),  # Optional: Add dropout to reduce overfitting
    layers.BatchNormalization(),  # Optional: Add batch normalization
    
    # Third hidden layer
    layers.Dense(units=512, activation='relu'),
    layers.Dropout(0.5),  # Optional: Add dropout to reduce overfitting
    layers.BatchNormalization(),  # Optional: Add batch normalization
    
    # Output layer for binary classification
    layers.Dense(units=1, activation='sigmoid')  # For binary classification, use 'sigmoid'
])
```

```python
model.compile(optimizer='adam', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])
```

```python
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)

history = model.fit(
    X_train, y,
    batch_size=4096,
    epochs=10,
    callbacks=[lr_scheduler]
)
```

```python
predictions = model.predict(X_test)
predictions_flat = predictions.reshape(-1)  # Ensure 1D
final = (predictions > 0.5).astype(int)
final_flat = final.reshape(-1)
```

```python
output = pd.DataFrame({
    'id': test_data['id'],  
    'Depression': final_flat
})

output.to_csv('mentalhealth_submission.csv', index=False)
```
